. Vérifier si deux listes sont égales.

```python
def empiler(liste,element):
    new_liste=[element]
    for i in liste:
         new_liste.append(i)
    return new_liste

ma_liste=[1,2,3,4]
print(empiler(ma_liste,5))

def enfiler(liste,element):
    new_liste=liste.copy()
    new_liste.append(element)
    return new_liste

ma_liste=[1,2,3,4]
print(enfiler(ma_liste,5))

def appartient(liste, element):
   if len(liste) == 0:
        return False
   if liste[0] == element:
        return True
  
   return appartient(liste[1:], element)
ma_liste = [1, 2, 3, 4, 5]
print(appartient(ma_liste, 0))  
print(appartient(ma_liste, 6))



def depiler(liste):
    if len(liste) == 0:
            return liste 
    premier_element = liste[0]
    new_liste = liste[1:]
    return premier_element,new_liste
    
    
ma_liste = [1, 2, 3, 4, 5]
element, new_liste = depiler( ma_liste)
print("element depile:", element)    
print("nouvelle liste:", new_liste)
  
    


def delete(liste,element):
    for i in range(len(liste)):
        if liste[i]==element:
            del liste[i]
            return liste
    
       
ma_liste = [1, 2, 3, 4, 5]
print("Liste avant suppression:", ma_liste) 
ma_liste = delete(ma_liste, 3)
print("Liste apres suppression de 3:", ma_liste) 


def longueur(liste):
    return len(liste)


ma_liste = [3, 1, 4, 1, 5, 9]
print("Longueur de la liste :", longueur(ma_liste))  

def remplacer(liste,element,newelement):
    for i in range(len(liste)):
        if liste[i]==element:
            liste[i]=newelement
            return liste
ma_liste=[1,2,3,4,5]
print("liste avant remplacement :",ma_liste)
ma_liste=remplacer(ma_liste,3,6)
print("liste apres remplacement :",ma_liste)
    
def afficher(liste):
    for i in range(len(liste)):
        print(liste[i], end=" " )
    print()
              
ma_liste=[1,2,3,4,5]
print("afficher la liste :")
afficher(ma_liste)

def occurence(liste,element):
    compteur=0
    for i in range(len(liste)):
        if liste[i]== element:
            compteur+=1
    return compteur
        
ma_liste=[1,2,3,4,2,6,2]
element_recherche=2
print(f"nombre de l occurence de {element_recherche} dans la liste est:",occurence(ma_liste,element_recherche))

def ordonnee(liste):
    for i in range(len(liste)-1):
        if liste[i] > liste[i+1]:
            return False
    
    return True
ma_liste=[1,2,3,4,5]
print(ordonnee(ma_liste))

def trier(liste):
    for i in range(len(liste)-1):
        for j in range(i+1,len(liste)):
            if liste[i]>liste[j]:
                liste[i],liste[j]=liste[j],liste[i]
    return liste
ma_liste=[5,2,8,1,9]
print(trier(ma_liste))

def insere_dans_liste_triee(liste,element):
    for i in range(len(liste)):
        if liste[i]>element:
            liste.insert(i,element)
            return liste
    
       
ma_liste=[1,2,5,8,9]
print(insere_dans_liste_triee(ma_liste,4))

def deux_listes(liste1,liste2):
    if len(liste1)!=len(liste2):
        return False
    for i in range(len(liste1)):
        if liste1[i]!=liste2[i]:
            return False
    return True
ma_liste1=[1,2,3,4,5]
ma_liste2=[1,2,3,4,5]
print(deux_listes(ma_liste1,ma_liste2))
ma_liste1=[1,2,3,4,5]
ma_liste2=[1,7,3,4,5]
print(deux_listes(ma_liste1,ma_liste2))

def concatener(liste1,liste2):
     new_liste = liste1[:] 
     for element in liste2:
         new_liste.append(element)
     return new_liste  
ma_liste1=[1,2,3,4,5]
ma_liste2=[6,7,8,9]
print(concatener(ma_liste1,ma_liste2))

def inverse(liste):
   return liste[::-1]

ma_liste=[1,2,3,4,5]
print(inverse(ma_liste))
```

Revoir l’exercice 5, en utilisant les listes et les fonctions pour saisir des nombres réels et pour calculer les fonctions suivantes (sans utiliser Numpy) :

. la somme

. la moyenne

. le produit

. le max

. le min

. l’écart-type

. la variance

. la covariance

Pour effectuer cet exercice, vous aurez besoin d'utiliser la fonction sqrt pour l'utilisation de la fonction racine carée
Pour cela, ajoutez en haut de l'exercice la ligne suivante :
from math import sqrt

```python
from math import sqrt

def somme(liste):
    return sum(liste)

def moyenne(liste):
    return sum(liste) / len(liste) if len(liste) > 0 else 0

def produit(liste):
    produit = 1
    for nombre in liste:
        produit *= nombre
    return produit

def calcul_max(liste):
    return max(liste)

def calcul_min(liste):
    return min(liste)

def variance(liste):
    moy = moyenne(liste)  
    return sum((x - moy) ** 2 for x in liste) / len(liste)

def ecart_type(liste):
    return sqrt(variance(liste))

def covariance(liste1, liste2):
    if len(liste1) != len(liste2):
        raise ValueError("Les deux listes doivent avoir la même longueur")

    moy1 = moyenne(liste1)
    moy2 = moyenne(liste2)

    return sum((liste1[i] - moy1) * (liste2[i] - moy2) for i in range(len(liste1))) / len(liste1)

nombres = [1.5, 4.2, 5.7, 6, 7.4]

print("Somme est:", somme(nombres))
print("Moyenne est :", moyenne(nombres))
print("Produit est :", produit(nombres))
print("Max est:", calcul_max(nombres))
print("Min est :", calcul_min(nombres))
print("Variance est :", variance(nombres))
print("Écart-type est:", ecart_type(nombres))

nombres2 = [2.3, 4.5, 1.8, 3.6, 5.0]
print("Covariance :", covariance(nombres, nombres2))
```

```python
def personne():
    Nom = input("Quel est votre nom ? ").strip()
    Prenom = input("Quel est votre prénom ? ").strip()
    Genre = input("Quel est votre genre ? ").strip()
    Age = int(input("Quel est votre âge ? ").strip())
    print("Nous avons le plaisir de vous présenter : ")
    print(Nom, "  ", Prenom, "  ", Genre, "  ", Age)
    return {"nom": Nom, "prenom": Prenom, "genre": Genre, "age": Age}

def lister_personnes(liste_personnes):
    print("\nListe des personnes :")
    for i, personne in enumerate(liste_personnes, 1):
        print(f"{i}. {personne['nom']} {personne['prenom']} ({personne['genre']}, {personne['age']} ans)")

def chercher_personne(liste_personnes, nom_recherche):
    nom_recherche = nom_recherche.strip().lower() 
    for personne in liste_personnes:
        if personne['nom'].strip().lower() == nom_recherche:
            print("\nPersonne trouvée :")
            print(f"{personne['nom']} {personne['prenom']} ({personne['genre']}, {personne['age']} ans)")
            return
    print("\nPersonne non trouvée.")
nb_personnes = int(input("Combien de personnes souhaitez-vous ajouter à la liste ? "))
liste_personnes = []

for _ in range(nb_personnes):
    liste_personnes.append(personne())

lister_personnes(liste_personnes)

nom_recherche = input("\nEntrez le nom de la personne à rechercher : ")
chercher_personne(liste_personnes, nom_recherche)
```
